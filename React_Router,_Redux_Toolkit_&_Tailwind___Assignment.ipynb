#Question 1 : Create a minimal router with two pages: Home (/) and About (/about). Show the links to move between pages.

- App.js

```
import React from "react";
import { BrowserRouter, Routes, Route, Link } from "react-router-dom";

function Home() {
  return (
    <div>
      <h2>Home Page</h2>
      <p>This is the home page of my small website.</p>
      <Link to="/about">Go to About Page</Link>
    </div>
  );
}

function About() {
  return (
    <div>
      <h2>About Page</h2>
      <p>This page gives some information.</p>
      <Link to="/">Go back to Home</Link>
    </div>
  );
}

function App() {
  return (
    <BrowserRouter>
      <div>
        <h1>Simple Router Example</h1>

        {/* Navigation Links */}
        <nav>
          <Link to="/">Home</Link> |
          <Link to="/about"> About</Link>
        </nav>

        <Routes>
          <Route path="/" element={<Home />} />
          <Route path="/about" element={<About />} />
        </Routes>

      </div>
    </BrowserRouter>
  );
}

export default App;

```

#Question 2 : After a user clicks “Save”, programmatically go to /thank-you

```
import React from "react";
import { BrowserRouter, Routes, Route, useNavigate } from "react-router-dom";

function FormPage() {
  const navigate = useNavigate();

  const handleSave = () => {
    // after saving data go to thank-you page
    navigate("/thank-you");
  };

  return (
    <div>
      <h2>Form Page</h2>
      <button onClick={handleSave}>Save</button>
    </div>
  );
}

function ThankYou() {
  return <h2>Thank you! Your data has been saved.</h2>;
}

function App() {
  return (
    <BrowserRouter>
      <Routes>
        <Route path="/" element={<FormPage />} />
        <Route path="/thank-you" element={<ThankYou />} />
      </Routes>
    </BrowserRouter>
  );
}

export default App;
```


#Question 5 : Only logged-in users should see /dashboard. Others should be redirected to /login. Write a minimal ProtectedRoute wrapper.

```
import React from "react";
import { BrowserRouter, Routes, Route, Navigate } from "react-router-dom";

const isLoggedIn = false; // change to true if user is logged in

function ProtectedRoute({ children }) {
  if (!isLoggedIn) {
    return <Navigate to="/login" />;
  }
  return children;
}

function Dashboard() {
  return <h2>Welcome to Dashboard</h2>;
}

function Login() {
  return <h2>Please Login First</h2>;
}

function App() {
  return (
    <BrowserRouter>
      <Routes>
        <Route
          path="/dashboard"
          element={
            <ProtectedRoute>
              <Dashboard />
            </ProtectedRoute>
          }
        />

        <Route path="/login" element={<Login />} />
      </Routes>
    </BrowserRouter>
  );
}

export default App;
```

#Question 6 : Create a counter slice with increment and decrement. Configure the store and provide it to React.

1. Counter Slice (counterSlice.js)
```
import { createSlice } from "@reduxjs/toolkit";

const counterSlice = createSlice({
  name: "counter",
  initialState: { value: 0 },

  reducers: {
    increment: (state) => {
      state.value += 1;
    },
    decrement: (state) => {
      state.value -= 1;
    },
  },
});

export const { increment, decrement } = counterSlice.actions;
export default counterSlice.reducer;
```

2. Store Configuration (store.js)
```
import { configureStore } from "@reduxjs/toolkit";
import counterReducer from "./counterSlice";

export const store = configureStore({
  reducer: {
    counter: counterReducer,
  },
});
```
3. Provide Store to React (main.js / index.js)
```
import React from "react";
import ReactDOM from "react-dom/client";
import { Provider } from "react-redux";
import App from "./App";
import { store } from "./store";

ReactDOM.createRoot(document.getElementById("root")).render(
  <Provider store={store}>
    <App />
  </Provider>
);

```

4. Use Counter in Component (App.js)

```
import React from "react";
import { useSelector, useDispatch } from "react-redux";
import { increment, decrement } from "./counterSlice";

function App() {
  const count = useSelector((state) => state.counter.value);
  const dispatch = useDispatch();

  return (
    <div>
      <h2>Counter: {count}</h2>

      <button onClick={() => dispatch(increment())}>Increment</button>
      <button onClick={() => dispatch(decrement())}>Decrement</button>
    </div>
  );
}

export default App;
```

```
import React from "react";
import { useSelector, useDispatch } from "react-redux";
import { increment, decrement } from "./counterSlice";

function Counter() {
  const count = useSelector((state) => state.counter.value);
  const dispatch = useDispatch();

  return (
    <div>
      <h2>Counter Value: {count}</h2>

      <button onClick={() => dispatch(increment())}>
        Increment
      </button>

      <button onClick={() => dispatch(decrement())}>
        Decrement
      </button>
    </div>
  );
}

export default Counter;
```


#Question 8 : Create a fetchProducts thunk that loads from /api/products. Handle loading and error in the slice.

```
import { createSlice, createAsyncThunk } from "@reduxjs/toolkit";

// thunk to fetch products
export const fetchProducts = createAsyncThunk(
  "products/fetchProducts",
  async () => {
    const response = await fetch("/api/products");
    const data = await response.json();
    return data;
  }
);

const productSlice = createSlice({
  name: "products",
  initialState: {
    items: [],
    loading: false,
    error: null,
  },

  reducers: {},

  extraReducers: (builder) => {
    builder
      .addCase(fetchProducts.pending, (state) => {
        state.loading = true;
        state.error = null;
      })
      .addCase(fetchProducts.fulfilled, (state, action) => {
        state.loading = false;
        state.items = action.payload;
      })
      .addCase(fetchProducts.rejected, (state, action) => {
        state.loading = false;
        state.error = "Failed to load products";
      });
  },
});

export default productSlice.reducer;
```

```
.product-card {
  @apply border p-4 rounded shadow transition duration-300 hover:shadow-lg;
}
```
- grid grid-cols-1 → shows 1 column on small screens.

- md:grid-cols-3 → shows 3 columns on medium screens and larger.

- gap-4 → adds space between items.

- transition and hover:shadow-lg → adds a hover effect.

- @apply is used to reuse Tailwind utility classes inside a custom class (product-card).
